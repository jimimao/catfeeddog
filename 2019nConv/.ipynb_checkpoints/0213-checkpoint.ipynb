```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
conv = pd.read_csv('./2019nCoV_national.csv')
index = pd.to_datetime(conv['Date'])
df = conv[['Infection','suspected','Recovery']]
df.index = index
df.fillna(method = "ffill",inplace = True)
df.fillna(0,inplace = True)
df.head()
```

```python
def power_func(x,a,b,c):
    return a*np.exp(b*x) + c
def linear_func(x,a,b):
    return a*x + b
```

```python
x = np.arange(1,df.shape[0]+1,1)
krgs_power,_ = curve_fit(power_func,x,df['Recovery'])
krgs_linear,_= curve_fit(linear_func,x,df['Infection'] + df['suspected'])
krgs_power,krgs_linear
```

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(x,power_func(x,*krgs_power),'b:',label = 'Recovery')
ax1.scatter(x,df['Recovery'],c= 'b',alpha = 0.4)
for a,b in zip(x,df['Recovery']):
    ax1.text(a,b-200,'{0:.0f}'.format(b),ha='center', va= 'bottom',fontsize=10)
# ax1.plot(x,linear_func(x,*krgs_linear),'r:',label = 'infection+suspected')
# ax1.scatter(x,df['Infection'] + df['suspected'],c= 'r',alpha = 0.4)
# for a,b in zip(x,df['Infection'] + df['suspected']):
#     ax1.text(a,b-200,'{0:.0f}'.format(b),ha='center', va= 'bottom',fontsize=10)
ax1.plot(x,df['Infection'])
plt.legend(loc = 'upper left')
```
